```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Users\22108160\Desktop\OIP (3).jpg")
cv2.imshow("Original Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#converting image into rgb
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.imshow("RGB Image", rgb_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#converting image into greyscale 
grayscale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray Image", grayscale_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#plotting histogram of original image 
plt.hist(img.ravel(), 256, [0, 256])
plt.title("Histogram of Original Image")
plt.show()
```

```python
#Contrast Stretching-Linear Transformation
import numpy as np 
min_val = np.min(img)
max_val = np.max(img)

#stretching pixel valyes
stretched_img = ((img - min_val)/(max_val - min_val))* 250
stretched_img = stretched_img.astype(np.uint8)

plt.imshow(stretched_img, cmap='gray')  
plt.title('conrast stretched image')
plt.show()
```

```python
#histogram equalization for gray image
equalized_img = cv2.equalizeHist(grayscale_img)
plt.imshow(equalized_img, cmap='gray')
plt.title("Histogram Equalized Image")
plt.show()

plt.hist(equalized_img.ravel(), 256, [0,256])
plt.title("Histogram of equalized Image")
plt.show()
```

```python
#clahe hist equalization
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_img = clahe.apply(grayscale_img)
plt.imshow(clahe_img, cmap = 'gray')
plt.title('CLAHE Image')
plt.show()
```

```python
#comparing results 
plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title('Original')

plt.subplot(1,3,2)
plt.imshow(stretched_img, cmap='gray')
plt.title('Contrast Stretched')

plt.subplot(1,3,3)
plt.imshow(clahe_img, cmap='gray')
plt.title('Clahe Enhanced')

plt.show()
```
