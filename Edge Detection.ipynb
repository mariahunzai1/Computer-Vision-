# Edge Detection 

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread(r"C:\Users\22108160\Desktop\images.jpg")
resized_img = cv2.resize(img, (400, 500))

img_rgb = cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(resized_img, cv2.COLOR_BGR2GRAY)
```

```python
# Sobel gradients
gradients_soblex = cv2.Sobel(img_gray, -1, 1, 0)
gradients_sobley = cv2.Sobel(img_gray, -1, 0, 1)
gradients_soblexy = cv2.addWeighted(gradients_soblex, 0.5, gradients_sobley, 0.5, 0)

# Laplacian gradient
gradients_laplacian = cv2.Laplacian(img_gray, -1)

# Canny edge detection
canny_output = cv2.Canny(img_gray, 80, 150)
```

```python
# Plotting  
plt.figure(figsize=(12, 12))

# Original image
plt.subplot(3, 3, 1)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')

# Sobel X Gradient
plt.subplot(3, 3, 2)
plt.imshow(gradients_soblex, cmap='gray')
plt.title('Sobel X')
plt.axis('off')

# Sobel Y Gradient
plt.subplot(3, 3, 3)
plt.imshow(gradients_sobley, cmap='gray')
plt.title('Sobel Y')
plt.axis('off')

# Sobel XY Gradient (combined X and Y)
plt.subplot(3, 3, 4)
plt.imshow(gradients_soblexy, cmap='gray')
plt.title('Sobel XY')
plt.axis('off')

# Laplacian Gradient
plt.subplot(3, 3, 5)
plt.imshow(gradients_laplacian, cmap='gray')
plt.title('Laplacian')
plt.axis('off')

# Canny edge detection output
plt.subplot(3, 3, 6)
plt.imshow(canny_output, cmap='gray')
plt.title('Canny Output')
plt.axis('off')

plt.tight_layout()
plt.show()
```
