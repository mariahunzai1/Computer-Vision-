```python
import cv2
from ultralytics import YOLO
```

```python
# Load pretrained YOLOv8 model
model = YOLO("yolov8n.pt")
```

```python
results = model("https://ultralytics.com/images/girl.jpg")
```

```python
import matplotlib.pyplot as plt
import cv2

img = results[0].plot()  # returns a BGR image with boxes
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```
